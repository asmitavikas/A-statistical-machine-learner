```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Step 1. Read data

```python
data = pd.read_csv('KMeansData') # 2 dimensional dataset
X = np.array(list(zip(data['V1'].values, data['V2'].values)))
```

```python
plt.scatter(data['V1'], data['V2'],  s=2);
```

#### Step 2. Define distance measure

```python
import numpy as np
# distance of each point in the cluster from the centroid
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

#### Step 3. Form clusters

```python
# form k random centroids for k clusters
k = 3
# X coordinates of random centroids
C_x = np.random.randint(0, np.max(X)-20, size=k)
# Y coordinates of random centroids
C_y = np.random.randint(0, np.max(X)-20, size=k)

C = np.array(list(zip(C_x, C_y)))
```

```python
print("Initial Centroids")
print(C)
```

```python
# Plotting along with the Centroids
plt.scatter(data['V1'], data['V2'], s=2)
plt.scatter(C_x, C_y, marker='*', s=200, c='black');
```

```python
# To store the value of centroids when it updates
C_old = np.zeros(C.shape)
```

```python
# Array to store cluster labels for each data point(0, 1, 2,..)
# this will tell us which data point is in what cluster
clusters = np.zeros(len(X))
```

```python
# Error func. - Distance between new centroids and old centroids
error = dist(C, C_old, None)
```

```python

# loop till error becomes zero or n_iter is complete

from copy import deepcopy
n, n_iter = 0, 8


while error != 0 or n < n_iter:
    # Assigning each data point to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i], C)
        cluster = np.argmin(distances)
        clusters[i] = cluster # assigning to closest cluster
    
    # Storing the old centroid values
    C_old = deepcopy(C)
    
    # Finding the new centroids by taking the average value in each cluster
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, C_old, None)
    n += 1
```

```python
colors = ['r', 'g', 'b', 'y', 'c', 'm']
fig, ax = plt.subplots()
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')

```
